# Upload model to pipeline demo



## Import

```python
import wallaroo
import wlauth
```

## Connect to Wallaroo

```python
# log in with a credential file
wl = wlauth.login('creds.json') # user: nina
```

## Connect to Wallaroo, upload model and deploy pipeline

```python
ccfraud_pp = wl.upload_model('ccfraud' , "./keras_ccfraud.onnx").deploy('ccfraud-pp')
```

```python
# Verify
ccfraud_pp.infer_from_file('./dev_smoke_test.json')
```

# Let's start streaming in transactions

```python
# !./start-sender.sh

```

# Let's replace the model

```python
model2 = wl.upload_model('ccfraud-xgb', './xgboost_ccfraud.onnx')
ccfraud_pp.replace_with_model_step(0, model2).deploy()
```
